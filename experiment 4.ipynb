## Q1) Factorial of a given number

```python
num = 5
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial:", fact)
```

## Q2) Check Armstrong number

```python
num = 153
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit**3
    temp = temp // 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

## Q3) Fibonacci series (first 5 terms)

```python
n = 5
a = 0
b = 1

print(a)
print(b)

for i in range(2, n):
    c = a + b
    print(c)
    a = b
    b = c
```

## Q4) Check prime number

```python
num = 7
flag = 0

for i in range(2, num):
    if num % i == 0:
        flag = 1
        break

if flag == 0:
    print("Prime number")
else:
    print("Not a prime number")
```

## Q5) Check palindrome number

```python
num = 121
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

## Q6) Sum of digits

```python
num = 1234
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum of digits:", sum)
```

## Q7) Numbers divisible by 5 or 7 (1 to 100)

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i)
        count = count + 1

print("Count:", count)
```

## Q9) Prime numbers between 1 and 100

```python
for num in range(2, 101):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```
