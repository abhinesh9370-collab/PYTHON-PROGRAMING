## 1. Check whether a number is divisible by 3 and 5 both

```python
num = 30

if num % 3 == 0 and num % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both 3 and 5")
```

## 2. Check whether a number is multiple of 5 or not

```python
num = 25

if num % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

## 3. Find the greatest among two numbers (or equal)

```python
a = 10
b = 10

if a > b:
    print("a is greater")
elif b > a:
    print("b is greater")
else:
    print("numbers are equal")
```

## 4. Find the greatest among three numbers (no two same)

```python
a = 12
b = 7
c = 9

if a > b and a > c:
    print("a is greatest")
elif b > a and b > c:
    print("b is greatest")
else:
    print("c is greatest")
```

## 5. Check quadratic roots (real or imaginary) and display roots

Equation: axÂ² + bx + c = 0

```python
a = 1
b = -5
c = 6

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    print("Real and different roots:", r1, r2)
elif d == 0:
    r = -b / (2*a)
    print("Real and equal roots:", r)
else:
    print("Imaginary roots")
```

## 6. Check whether a year is a leap year 

```python
year = 2024

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

## Q8. Grade Sheet Program

```python
name = "Rohit Sharma"
roll = "R17234512"
sapid = "50005673"
sem = 1
course = "B.Tech CSE AI & ML"

pds = 70
python = 80
chemistry = 90
english = 60
physics = 50

total = pds + python + chemistry + english + physics
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O"

print("Name:", name)
print("Roll Number:", roll, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chemistry)
print("English:", english)
print("Physics:", physics)
print("Percentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
