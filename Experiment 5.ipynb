### Experiment 5: String and Sets 

## 1) Count and display number of capital letters in a string

```python
s = "Hello PYthon WORLD"
count = 0

for ch in s:
    if ch.isupper():
        count += 1

print("String:", s)
print("Capital letters:", count)
```

## 2) Count total number of vowels in a string

```python
s = "Education is Important"
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("String:", s)
print("Total vowels:", count)
```

## 3) Print words in separate lines

```python
sentence = "Python is very easy"
words = sentence.split()

for word in words:
    print(word)
```

## 4) Count substring occurrences (left to right)

```python
string = "ABCDCDC"
substring = "CDC"

count = 0
i = 0

while i <= len(string) - len(substring):
    if string[i:i+len(substring)] == substring:
        count += 1
    i += 1

print("String:", string)
print("Substring:", substring)
print("Occurrences:", count)
```

## 5) Count occurrences of each alphabet (case-insensitive)

```python
s = "ABaBCbGc"
s = s.upper()

count = {}

for ch in s:
    if ch.isalpha():
        count[ch] = count.get(ch, 0) + 1

for key in sorted(count):
    print(count[key], key)
```

## 6) Count unique words using set

```python
sentence = "Python is easy and Python is powerful"
words = sentence.split()

unique_words = set(words)

print("Unique words:", unique_words)
print("Count of unique words:", len(unique_words))
```

## 7) Set operations on fruits 

```python
s1 = {"apple", "banana", "mango"}
s2 = {"banana", "orange", "mango"}

# a) Fruits in both sets
print("Common fruits:", s1 & s2)

# b) Fruits only in s1
print("Only in s1:", s1 - s2)

# c) Count of all fruits
all_fruits = s1 | s2
print("Total fruits count:", len(all_fruits))
```

## 8.	Take two sets and apply various set operations on them: 
S1 = {Red, yellow, orange, blue} 
S2 = {violet, blue, purple}


```python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)

print("Intersection:", S1 & S2)

print("Difference (S1 - S2):", S1 - S2)

print("Difference (S2 - S1):", S2 - S1)

print("Symmetric Difference:", S1 ^ S2)
```
