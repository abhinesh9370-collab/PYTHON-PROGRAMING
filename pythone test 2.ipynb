```python
class Employee:
    company = "abc Ltd"

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print("Employee name:", self.name)
        print("Employee salary:", self.salary)
        print("Employee company:", self.company)

    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company


e1 = Employee("Aman", 50000)
e2 = Employee("Riya", 60000)

e1.show_details()
e2.show_details()

Employee.change_company("xyz Ltd")

e1.show_details()
e2.show_details()
```

```python
class MathHelper:

    @staticmethod
    def is_even(num):
        if num % 2 == 0:
            return True
        else:
            return False

print(MathHelper.is_even(8))
print(MathHelper.is_even(11))
```
