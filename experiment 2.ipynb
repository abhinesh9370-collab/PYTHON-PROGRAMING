## 6. Hypotenuse using Pythagoras theorem

```python
a = 3   # base
b = 4   # height

c = (a*a + b*b) ** 0.5

print("Hypotenuse:", c)
```

## 8. Area of Triangle (three sides given)

```python
a = 5
b = 6
c = 7

s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of triangle:", area)
```

## 9. Convert seconds to hours, minutes and seconds

```python
total_seconds = 4520

hours = total_seconds // 3600
remaining_seconds = total_seconds % 3600
minutes = remaining_seconds // 60
seconds = remaining_seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

## 10. Swap two numbers without extra variable

```python
a = 5
b = 10

a = a + b
b = a - b
a = a - b

print("a =", a)
print("b =", b)
```

## 11. Sum of first n natural numbers

```python
n = 10

sum = n * (n + 1) // 2

print("Sum of first", n, "natural numbers:", sum)
```

## 12. Truth table for bitwise operators

```python
a = 1
b = 0

print("a & b =", a & b)
print("a | b =", a | b)
print("a ^ b =", a ^ b)
```

## 13. Left shift and Right shift

```python
num = 8

print("Number:", num)
print("Left Shift (num << 1):", num << 1)
print("Right Shift (num >> 1):", num >> 1)
```

## Q14: Membership Operator

```python
sequence = (10, 20, 56, 78, 89)

num = 56

if num in sequence:
    print(num, "is present in the sequence")
else:
    print(num, "is not present in the sequence")
```
