## 1️ Function to find maximum and minimum (no built-in)

```python
def find_max_min(numbers):
    max_val = numbers[0]
    min_val = numbers[0]

    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val
```

## 2. Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total
```

## 3️. Print 1 to n using recursion (no loop)

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)
```

## 4️ Recursive function to print Fibonacci series up to n terms

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

def print_fibonacci(n):
    for i in range(n):
        print(fibonacci(i), end=" ")
```

## 5️ Lambda function to find volume of cone

```python
import math

volume_of_cone = lambda r, h: (1/3) * math.pi * r * r * h
```

c) Variable Length Argument

```python
def total_sum(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum:", total)

total_sum(10, 20, 30)
total_sum(5, 15)
```
