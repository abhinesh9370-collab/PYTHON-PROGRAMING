## Section 1: Creating a Student Class and Objects

Create a Student class with attributes name, sap_id, and marks (physics, chemistry, mathematics). Create 3 objects by taking inputs from the user and display details of all students.

```python
# Section 1: Simple Student Class
class Student:
    """A class to represent a student"""
    
    def __init__(self, name, sap_id, physics, chemistry, maths):
        self.name = name
        self.sap_id = sap_id
        self.physics = physics
        self.chemistry = chemistry
        self.maths = maths
    
    def display(self):
        """Display student details"""
        print(f"Name: {self.name}")
        print(f"SAP ID: {self.sap_id}")
        print(f"Physics: {self.physics}")
        print(f"Chemistry: {self.chemistry}")
        print(f"Maths: {self.maths}")
        print()

# Create 3 student objects
print("=" * 50)
print("CREATING 3 STUDENT OBJECTS")
print("=" * 50)

students = []
for i in range(3):
    print(f"\nEnter details for Student {i+1}:")
    name = input("Enter name: ")
    sap_id = input("Enter SAP ID: ")
    physics = float(input("Enter Physics marks: "))
    chemistry = float(input("Enter Chemistry marks: "))
    maths = float(input("Enter Maths marks: "))
    
    student = Student(name, sap_id, physics, chemistry, maths)
    students.append(student)

# Display all student details
print("\n" + "=" * 50)
print("DISPLAYING ALL STUDENTS")
print("=" * 50)
for i, student in enumerate(students):
    print(f"\nStudent {i+1}:")
    student.display()
```

## Section 2: Enhanced Student Class with Methods

Add a constructor to initialize student details for n students and implement:
- **Display()** - Display student details
- **Find_Marks_percentage()** - Calculate percentage for each student
- **Display_result()** - Show pass/fail (Pass if marks in each subject > 40%, else Fail)
- **Class_average()** - Find average percentage of the class

```python
# Section 2: Enhanced Student Class with Methods
class EnhancedStudent:
    """Enhanced Student class with methods for percentage, result, and class average"""
    
    students_list = []  # Class variable to store all students
    
    def __init__(self, name, sap_id, physics, chemistry, maths):
        """Constructor to initialize student details"""
        self.name = name
        self.sap_id = sap_id
        self.physics = physics
        self.chemistry = chemistry
        self.maths = maths
        EnhancedStudent.students_list.append(self)
    
    def display(self):
        """Display student details"""
        print(f"Name: {self.name}")
        print(f"SAP ID: {self.sap_id}")
        print(f"Physics: {self.physics}")
        print(f"Chemistry: {self.chemistry}")
        print(f"Maths: {self.maths}")
    
    def find_marks_percentage(self):
        """Calculate percentage of the student"""
        total = self.physics + self.chemistry + self.maths
        percentage = (total / 300) * 100  # Out of 100 per subject
        return percentage
    
    def display_result(self):
        """Display pass/fail result based on marks in each subject"""
        percentage = self.find_marks_percentage()
        
        # Check if marks in each subject > 40%
        phy_marks = (self.physics / 100) * 100
        chem_marks = (self.chemistry / 100) * 100
        math_marks = (self.maths / 100) * 100
        
        if self.physics >= 40 and self.chemistry >= 40 and self.maths >= 40:
            result = "PASS"
        else:
            result = "FAIL"
        
        print(f"Overall Percentage: {percentage:.2f}%")
        print(f"Result: {result}")
    
    @classmethod
    def class_average(cls):
        """Find average percentage of the entire class"""
        if len(cls.students_list) == 0:
            return 0
        
        total_percentage = sum(student.find_marks_percentage() for student in cls.students_list)
        average = total_percentage / len(cls.students_list)
        return average

# Create student objects
print("=" * 50)
print("ENHANCED STUDENT CLASS DEMONSTRATION")
print("=" * 50)

# Clear the students list
EnhancedStudent.students_list = []

# Create students
print("\nEnter details for n students:")
n = int(input("Enter number of students: "))

for i in range(n):
    print(f"\nEnter details for Student {i+1}:")
    name = input("Enter name: ")
    sap_id = input("Enter SAP ID: ")
    physics = float(input("Enter Physics marks (out of 100): "))
    chemistry = float(input("Enter Chemistry marks (out of 100): "))
    maths = float(input("Enter Maths marks (out of 100): "))
    
    student = EnhancedStudent(name, sap_id, physics, chemistry, maths)

# Display all students with their details
print("\n" + "=" * 50)
print("DISPLAYING ALL STUDENTS WITH DETAILS")
print("=" * 50)

for i, student in enumerate(EnhancedStudent.students_list):
    print(f"\n--- Student {i+1} ---")
    student.display()
    print(f"Percentage: {student.find_marks_percentage():.2f}%")
    print("Result: ", end="")
    student.display_result()

# Display class average
print("\n" + "=" * 50)
print(f"CLASS AVERAGE PERCENTAGE: {EnhancedStudent.class_average():.2f}%")
print("=" * 50)
```

## Section 3: Different Types of Inheritance

Demonstrate different types of inheritance:
1. **Single Inheritance** - One child class inherits from one parent class
2. **Multiple Inheritance** - One child class inherits from multiple parent classes
3. **Multilevel Inheritance** - Chain of inheritance (grandparent -> parent -> child)
4. **Hierarchical Inheritance** - Multiple child classes inherit from one parent class

```python
# Section 3: Different Types of Inheritance

print("\n" + "=" * 50)
print("SECTION 3: DIFFERENT TYPES OF INHERITANCE")
print("=" * 50)

# 1. SINGLE INHERITANCE
print("\n1. SINGLE INHERITANCE")
print("-" * 50)

class Animal:
    """Parent class"""
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound"

class Dog(Animal):
    """Child class inherits from Animal"""
    def speak(self):
        return f"{self.name} barks"

dog = Dog("Buddy")
print(dog.speak())

# 2. MULTIPLE INHERITANCE
print("\n2. MULTIPLE INHERITANCE")
print("-" * 50)

class FlyingAbility:
    """First parent class"""
    def fly(self):
        return "This creature can fly"

class SwimmingAbility:
    """Second parent class"""
    def swim(self):
        return "This creature can swim"

class Duck(FlyingAbility, SwimmingAbility):
    """Child class inherits from two parent classes"""
    def __init__(self, name):
        self.name = name

duck = Duck("Donald")
print(f"Duck name: {duck.name}")
print(duck.fly())
print(duck.swim())

# 3. MULTILEVEL INHERITANCE
print("\n3. MULTILEVEL INHERITANCE")
print("-" * 50)

class Vehicle:
    """Grandparent class"""
    def __init__(self, brand):
        self.brand = brand
    
    def show_brand(self):
        return f"Brand: {self.brand}"

class Car(Vehicle):
    """Parent class inherits from Vehicle"""
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    
    def show_model(self):
        return f"Model: {self.model}"

class ElectricCar(Car):
    """Child class inherits from Car"""
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
    
    def show_battery(self):
        return f"Battery Capacity: {self.battery_capacity} kWh"

e_car = ElectricCar("Tesla", "Model 3", 75)
print(e_car.show_brand())
print(e_car.show_model())
print(e_car.show_battery())

# 4. HIERARCHICAL INHERITANCE
print("\n4. HIERARCHICAL INHERITANCE")
print("-" * 50)

class Shape:
    """Parent class"""
    def __init__(self, color):
        self.color = color
    
    def show_color(self):
        return f"Color: {self.color}"

class Circle(Shape):
    """First child class"""
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    """Second child class"""
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

circle = Circle("Red", 5)
rectangle = Rectangle("Blue", 4, 6)

print(f"Circle - {circle.show_color()}, Area: {circle.area()}")
print(f"Rectangle - {rectangle.show_color()}, Area: {rectangle.area()}")
```

## Section 4: Method Overriding

Method Overriding allows a subclass to provide a specific implementation of a method that is already defined in its superclass.

```python
# Section 4: Method Overriding

print("\n" + "=" * 50)
print("SECTION 4: METHOD OVERRIDING")
print("=" * 50)

class Parent:
    """Parent class with a method"""
    def display(self):
        print("This is the parent class method")
    
    def greet(self):
        return "Hello from Parent class"

class Child(Parent):
    """Child class overrides the parent's method"""
    def display(self):
        print("This is the child class method (OVERRIDDEN)")
    
    def greet(self):
        return "Hello from Child class"

# Create instances
parent_obj = Parent()
child_obj = Child()

print("\nParent class method call:")
parent_obj.display()
print(parent_obj.greet())

print("\nChild class method call (Overridden):")
child_obj.display()
print(child_obj.greet())

# Another example with Employee hierarchy
print("\n" + "-" * 50)
print("Example: Employee and Developer")
print("-" * 50)

class Employee:
    """Parent class for employees"""
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def calculate_bonus(self):
        """Parent method"""
        return self.salary * 0.10  # 10% bonus

class Developer(Employee):
    """Developer class overrides the bonus calculation"""
    def calculate_bonus(self):
        """Overridden method for developers"""
        return self.salary * 0.15  # 15% bonus (higher for developers)

class Manager(Employee):
    """Manager class overrides the bonus calculation"""
    def calculate_bonus(self):
        """Overridden method for managers"""
        return self.salary * 0.20  # 20% bonus (higher for managers)

# Create instances
emp = Employee("John", 50000)
dev = Developer("Alice", 60000)
mgr = Manager("Bob", 70000)

print(f"\nEmployee {emp.name}:")
print(f"Salary: ${emp.salary}")
print(f"Bonus: ${emp.calculate_bonus()}")

print(f"\nDeveloper {dev.name}:")
print(f"Salary: ${dev.salary}")
print(f"Bonus: ${dev.calculate_bonus()}")

print(f"\nManager {mgr.name}:")
print(f"Salary: ${mgr.salary}")
print(f"Bonus: ${mgr.calculate_bonus()}")
```

## Section 5: Operator Overloading

Create a Point class with x and y values, and implement operator overloading to add two Point objects.
- Example: If P1(x=10, y=20) and P2(x=12, y=15), then P3 = P1 + P2 results in P3(x=22, y=35)

```python
# Section 5: Operator Overloading

print("\n" + "=" * 50)
print("SECTION 5: OPERATOR OVERLOADING")
print("=" * 50)

class Point:
    """Point class with x and y coordinates"""
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """Overload the + operator to add two Point objects"""
        return Point(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Overload the - operator to subtract two Point objects"""
        return Point(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """Overload the * operator to multiply a Point by a scalar"""
        return Point(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        """Overload the == operator to compare two Point objects"""
        return self.x == other.x and self.y == other.y
    
    def __str__(self):
        """String representation of Point"""
        return f"Point(x={self.x}, y={self.y})"
    
    def __repr__(self):
        """Official representation of Point"""
        return f"Point({self.x}, {self.y})"

# Test operator overloading
print("\n--- Testing Operator Overloading with Point Class ---\n")

# Create two point objects
P1 = Point(10, 20)
P2 = Point(12, 15)

print(f"P1 = {P1}")
print(f"P2 = {P2}")

# Addition of two points
P3 = P1 + P2
print(f"\nP3 = P1 + P2")
print(f"P3 = {P3}")
print(f"Expected: Point(x=22, y=35)")

# Subtraction of two points
P4 = P1 - P2
print(f"\nP4 = P1 - P2")
print(f"P4 = {P4}")

# Multiplication with scalar
P5 = P1 * 2
print(f"\nP5 = P1 * 2")
print(f"P5 = {P5}")

# Comparison of two points
P6 = Point(10, 20)
P7 = Point(10, 20)
P8 = Point(22, 35)

print(f"\nComparison:")
print(f"P6 = {P6}")
print(f"P7 = {P7}")
print(f"P8 = {P8}")
print(f"P6 == P7: {P6 == P7}")
print(f"P6 == P8: {P6 == P8}")

# More examples
print("\n" + "-" * 50)
print("Additional Examples:")
print("-" * 50)

P9 = Point(5, 10)
P10 = Point(3, 7)
P11 = P9 + P10

print(f"\nP9 = {P9}")
print(f"P10 = {P10}")
print(f"P9 + P10 = {P11}")
print(f"Expected: Point(x=8, y=17)")

P12 = P9 * 3
print(f"\nP9 * 3 = {P12}")
print(f"Expected: Point(x=15, y=30)")
```
