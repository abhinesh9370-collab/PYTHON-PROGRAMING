```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] =+ 1
    return d
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return total - sum(nums)

nums = [3, 0, 1]
print(missing_number(nums))
```

## Q5

```python
def moving_average(stream, w):
    result = []
    window_sum = 0

    for i in range(len(stream)):
        window_sum += stream[i]
        if i >= w:
            window_sum -= stream[i - w]
        avg = window_sum / min(i + 1, w)
        result.append(avg)

    return result
```

```python
def rle_decode(s):
    result = []
    i = 0
    n = len(s)

    while i < n:
        if not s[i].isdigit():
            raise ValueError("Invalid format: count expected")

        count = 0
        while i < n and s[i].isdigit():
            count = count * 10 + int(s[i])
            i += 1

        if i >= n:
            raise ValueError("Invalid format: missing character")
        if s[i] == "\\":
            i += 1
            if i >= n:
                raise ValueError("Invalid escape sequence")
            char = s[i]
        else:
            char = s[i]
        result.append(char * count)
        i += 1

    return "".join(result)
```
